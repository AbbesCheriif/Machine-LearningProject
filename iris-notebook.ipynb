```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
```

```python
# Charger les données de l'utilisateur (vous devez implémenter cette fonction)
def user_input_features():
    # Exemple de données fictives
    data = [[5.1, 3.5, 1.4, 0.2], [6.2, 3.4, 5.4, 2.3]]
    df = pd.DataFrame(data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
    return df
```

```python
# Charger le dataset Iris
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
```

```python
iris_df.info()
```

```python
# Créer un classificateur RandomForest
clf = RandomForestClassifier()
clf.fit(iris.data, iris.target)
```

```python
# Obtenir les données d'entrée utilisateur
df = user_input_features()

# Prédire les classes et les probabilités
prediction = clf.predict(df)
prediction_proba = clf.predict_proba(df)
```

```python
# Afficher les résultats de la prédiction
print("Résultats de la prédiction :")
for i in range(len(prediction)):
    print(f"Échantillon {i+1} - Classe prédite : {iris.target_names[prediction[i]]}, Probabilités : {prediction_proba[i]}")
```

```python
# Visualisation des caractéristiques du dataset Iris
plt.figure(figsize=(10, 8))
for column in iris_df.columns[:-1]:
    plt.hist(iris_df[column], bins=20, alpha=0.5, label=column)
plt.xlabel('Valeurs')
plt.ylabel('Fréquence')
plt.title('Distribution des caractéristiques dans le dataset Iris')
plt.legend()
plt.show()
```
